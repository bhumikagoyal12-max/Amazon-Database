```python
import pandas as pd

df = pd.read_csv("/kaggle/input/datasets/tarkkaanko/amazon/amazon_reviews.csv")

print(df.head())
```

```python
import nltk
text=df['reviewText']
text=text.str.lower()
print(text)
```

```python
#tokenize into word or sentence
from nltk.tokenize import sent_tokenize,word_tokenize
df = df[df["reviewText"].notna()]
df["sentences"] = df["reviewText"].apply(sent_tokenize)
df["words"] = df["reviewText"].apply(word_tokenize)
print(df["sentences"])
print(df["words"])
```

```python
#remove punctuation marks
df['reviewText']=df['reviewText'].str.replace('[^\w\s]', '')
print(df['reviewText'])
```

```python
#stop word removal word doesnt contain meaning
from nltk.corpus import stopwords
my_stopwords={"a","an","the"}

nltk.download("stopwords")
nltk_stopwords = set(stopwords.words("english"))
nltk_stopwords.add(".")
nltk_stopwords.discard("but")
clean_list = []
for sentence in (df['reviewText']):
       sentence = str(sentence).lower()
       words = sentence.split()
       new_words = []
       for word in words:
        if word not in nltk_stopwords:
            new_words.append(word)
    
       clean_sentence = " ".join(new_words)
       clean_list.append(clean_sentence)
df['clean_text'] = clean_list

print(df[['reviewText','clean_text']].head())   
```

```python
#convert word into stopwords
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
text3=[stemming.stem(word) for word in words]
print(text3)
```

```python
#POS tagging
from nltk.tag import DefaultTagger
import nltk

nltk.download('averaged_perceptron_tagger_eng')
py_tag = DefaultTagger('NN')
tag_txt = py_tag.tag(text3)
print("Self tagging:\n",tag_txt)
print("Tagged by Tagger:\n",nltk.pos_tag(text3))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder,RobustScaler,StandardScaler
from sklearn.base import BaseEstimator,TransformerMixin
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
#split the data
X=df.drop(columns=['day_diff'])
y=df['day_diff']

#feature selection
numeric_features=X.select_dtypes(include=["int64","float64"]).columns.tolist()
cat_feature=df.select_dtypes(exclude=["int64","float64"]).columns.tolist()

```

```python
#preprocessing
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
numeric_transformer=Pipeline(steps=[("imputer",SimpleImputer(strategy="median")),
                                    ('scaler',StandardScaler())])
cat_transformer=Pipeline(steps=[("imputer",SimpleImputer(strategy="most_frequent")),
                             ("onehot",OneHotEncoder(handle_unknown="ignore"))])

preprocess=ColumnTransformer(transformers=[("num",numeric_transformer,numeric_features),
                                          ("cat",cat_transformer,cat_feature)])
```

```python
#train and test the data

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
X_train = X_train.loc[:, X_train.columns.str.strip()]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

num_cols = [col for col in ["day_diff","total_vote","score_pos_neg_diff","score_average_rating","wilson_lower_bound"] if col in X_train.columns]

preprocessor = ColumnTransformer([
    ("num", StandardScaler(), num_cols),
    ...
])
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import(accuracy_score,precision_score,f1_score,roc_auc_score,recall_score,confusion_matrix)
```

```python
#model
models={
    "Logistic Regression":LogisticRegression(),
    "KNN Classifier":KNeighborsClassifier(n_neighbors=7),
    "SVM Classifier":SVC(random_state=100),
    "Decision Tree":DecisionTreeClassifier(random_state=100),
    "Random Forest":RandomForestClassifier(random_state=100)
    
}
```

```python
#evaluation of model
from sklearn.metrics import classification_report
result=[]
for name, model in models.items():
    pipeline = Pipeline(steps=[
   
    ("preprocessor", preprocess),
    ("model", SVC(probability=True))
])
    pipeline.fit(X_train,y_train)
    y_pred=pipeline.predict(X_test)

    acc=accuracy_score(y_pred,y_test)
    f1=f1_score(y_test,y_pred,average="weighted")
    pre=precision_score(y_test,y_pred,average="weighted")
    rec=recall_score(y_test,y_pred,average="weighted")
   
    result.append([name,acc,f1,pre,rec])
    print("="*100)
    print("model name",model)
    print("="*100)
    print("accuracy",round(acc,4))
    print("f1",round(f1,4))
    print("precision",round(pre,4))
    print("recall",round(rec,4))
   
    print("confusion matrix \n",confusion_matrix(y_pred,y_test))
    print("classification report \n",classification_report(y_test,y_pred))


```
